```python
#Andre Barajas
#Spring 2020
#Project 3
import matplotlib.pyplot as plt 
import numpy as py
from scipy.stats import binom  as bino
from scipy.stats import poisson as poi
#problem 0
def expermintal_bernoulli_trials(c, n, p, N):
    #random variable X 
    X = py.zeros((N, 1))
    for k in range(0, N):
        success = 0
        die0 = py.random.choice(c, n, p=p)
        die1 = py.random.choice(c, n, p=p)
        die2 = py.random.choice(c, n, p=p)
        #attempt three dice n rolls
        for j in range(0,n):
            #if 1st =1, 2nd =2 and 3rd = 3 then success
            if die0[j] == 1 and die1[j] == 2 and die2[j] == 3:
                success += 1     
        X[k] = success    
    graph_b = range(0, 20)
    sb = py.size(graph_b)
    h1, bin_edges = py.histogram(X, bins = graph_b)
    b1 = bin_edges[0: sb - 1]
    plt.close("all")
    prob = h1 / N
    #plot histogram
    plt.stem(b1, prob) 
    plt.title("Bernoulli Trials: PMF - Simulation Results")
    plt.xlabel("# Successes in n trials")
    plt.ylabel("Probability of X Variable")
#problem 1
def binomial_distribution(n, p_success):
    #p of success in a single roll
    
    q_failure = 1 - p_success
    #simulate
    i = py.arange(0,20)
    #using built int binomial python function for probability
    bino_output = bino.pmf(i, n, p_success)
    plt.stem(i, bino_output)
    plt.title("Bernoulli Trials: PMF - Binomial Formula")
    plt.xlabel('Number of Successes in n: 1000 trials')
    plt.ylabel('Probability')
    #plot historgram
    plt.show()
#problem 2
def poisson_distribution(n_attemps, p_success):
    #where / = np 
    lamda = n_attemps * p_success
    j = py.arange(0, 20)
    poisson_output = poi.pmf(j, lamda)
    plt.stem(j, poisson_output)
    #graph historgram
    plt.title("Bernoulli Trials: PMF - Poisson Approximation")
    plt.xlabel('Number of Successes in n: 1000 trials')
    plt.ylabel('Probability')
    plt.show


n = 1000
c = [1,2,3,4,5,6]
N = 10000
p = [0.2,  0.1,  0.15, 0.3, 0.2, 0.05]
    
expermintal_bernoulli_trials(c, n, p, N)
    



```
